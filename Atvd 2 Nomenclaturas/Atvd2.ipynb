```python
##Código 02

def find_largest_number(numbers):
    largest_number = numbers[0]

    for number in numbers:
        if number > largest_number:
            largest_number = number

    return largest_number


largest = find_largest_number([3, 7, 2, 9, 4])
print("Maior número:", largest)
```

```python
##Código 05

def find_largest_even_number(numbers):
    largest_even_number = None

    for number in numbers:
        if number % 2 == 0:
            if largest_even_number is None or number > largest_even_number:
                largest_even_number = number

    return largest_even_number


result = find_largest_even_number([3, 10, 7, 8, 15])
print(result)
```

```python
##Código 07

def convert_celsius_to_fahrenheit(celsius_temperature):
    return (celsius_temperature * 9 / 5) + 32

def convert_fahrenheit_to_celsius(fahrenheit_temperature):
    return (fahrenheit_temperature - 32) * 5 / 9

celsius = 25
fahrenheit = 77

print(convert_celsius_to_fahrenheit(celsius))
print(convert_fahrenheit_to_celsius(fahrenheit))
```

```python
##Código 08

def calculate_average(grades):
    return sum(grades) / len(grades)


def is_student_approved(grades, passing_score=60):
    average = calculate_average(grades)
    return average >= passing_score


student_grades = [70, 80, 50]

average = calculate_average(student_grades)
approved = is_student_approved(student_grades)

print("Média:", average)
print("Aprovado:", approved)
```

```python
##Código 09

class BankAccount:
    def __init__(self, account_holder, initial_balance):
        self.account_holder = account_holder
        self.balance = initial_balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Saldo insuficiente")

    def get_balance(self):
        return self.balance

    def get_account_summary(self):
        return f"Conta: {self.account_holder}, Saldo: {self.balance}"


def calculate_total_balance(accounts):
    return sum(account.balance for account in accounts)


def find_account_with_highest_balance(accounts):
    return max(accounts, key=lambda account: account.balance)


# Instâncias
alice_account = BankAccount("Alice", 500)
bob_account = BankAccount("Bob", 1200)
charlie_account = BankAccount("Charlie", 700)

# Operações
alice_account.deposit(200)
bob_account.withdraw(200)

print(alice_account.get_balance())
print(bob_account.get_account_summary())

accounts = [alice_account, bob_account, charlie_account]

print("Saldo total:", calculate_total_balance(accounts))
print("Maior saldo:", find_account_with_highest_balance(accounts).account_holder)
```
